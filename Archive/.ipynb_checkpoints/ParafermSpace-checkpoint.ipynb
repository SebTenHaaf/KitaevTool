# Majorana solution

## Two-operators, real terms

```python
## Full operator basis has 16 vectors
all_seqs = [1, c_up, c_down, a_up, a_down, 
            c_up*a_up, c_down*a_down, c_up*a_down, c_down*a_up, 
            c_down*c_up,a_down*a_up,
            c_up*a_down*a_up, c_down*a_down*a_up, c_down*c_up*a_down, c_down*c_up*a_up,  
            c_up*c_down*a_up*a_down]

N_basis_vecs = len(all_seqs)
```

```python
## Generate initial Operator Sequence and variables
full_vector = fst.OperSequence()
a_vec = sp.Matrix(N_basis_vecs, 1, lambda i, _: 0)
identity_vec = sp.Matrix(N_basis_vecs, 1, lambda i, _: 0)
identity_vec[0]=1

real_terms = [5,6,15]

for i,seq in enumerate(all_seqs):
    if i in real_terms:
        a = sp.Symbol(f'a_{i}', real=True, positive=False,commutative=True)
    else:
        a = sp.Symbol(f'a_{i}', real=True, positive=False,commutative=True)
    if i==0 or i>10:
        a=0
    a_vec[i]=a
    full_vector += a*seq

all_symbols = [a_vec[i] for i in range(len(a_vec))]
```

```python
## Get conditions for Hermitian operator
conj_vector = ~full_vector
conj_vector.normal_order()
a_vec_conj = sp.Matrix(N_basis_vecs, 1, lambda i, _: 0)
for idx in range(len(conj_vector.oper_list)):
    try:
        basis_idx = vector_basis.index(conj_vector.oper_list[idx])
    except:
        continue
    a_vec_conj[basis_idx] += conj_vector.weights[idx]

## Constraints on the Hermition-ness of the vector
res = sp.solve(a_vec-a_vec_conj)
for key in res:
    a_vec = a_vec.subs(key,res[key])

hermitian_vector = fst.OperSequence()
for i,seq in enumerate(all_seqs):
    hermitian_vector += a_vec[i]*seq
```

```python
## Obtain the Matrix form of the action of the Hermitian vector
vector_basis = full_vector.oper_list
M = np.zeros((N_basis_vecs,N_basis_vecs))
M = sp.Matrix(N_basis_vecs, N_basis_vecs, lambda i, j: 0)

for col_idx, basis_oper in enumerate(all_seqs):
    vec_outp = hermitian_vector*basis_oper
    for idx in range(len(vec_outp.oper_list)):
        try:
            basis_idx = vector_basis.index(vec_outp.oper_list[idx])
        except:
            continue
        M[basis_idx,col_idx] += vec_outp.weights[idx]
```

```python
## Set up system of equations for v^2=I
to_solve_basis = M@a_vec - identity_vec
to_solve_basis
```

```python
M**2*identity_vec - identity_vec
```

```python
## Substitute the spherical parameterization for term [0]
theta1, theta2 = sp.symbols('theta1 theta2', real=True)
theta3, theta4 = sp.symbols('theta3 theta4', real=True)

solve_a3 = sp.sin(theta1)*sp.cos(theta2)
solve_a4 = sp.sin(theta1)*sp.sin(theta2)
solve_a9 = sp.cos(theta1)
to_solve = to_solve_basis.subs(all_symbols[3], solve_a3).subs(all_symbols[4], solve_a4).subs(all_symbols[9], solve_a9)
to_solve.simplify()

## Solve remaining linear equations
solve_a5 = sp.solve(to_solve[1], all_symbols[5])[0]
to_solve = to_solve.subs(all_symbols[5], solve_a5)
solve_a6 = sp.solve(to_solve[2], all_symbols[6])[0]
to_solve = to_solve.subs(all_symbols[6], solve_a6)
to_solve.simplify()
solve_a8 = sp.solve(to_solve[14], all_symbols[8])[0]
to_solve = to_solve.subs(all_symbols[8], solve_a8)
to_solve.simplify()
solve_a6 = solve_a6.subs(all_symbols[8], solve_a8).simplify()
solve_a5 = solve_a5.subs(all_symbols[8], solve_a8).simplify()
```

```python
## Substitute solutions into original vector
a_vec_solved = a_vec.subs(all_symbols[3],solve_a3).subs(all_symbols[4],solve_a4).subs(all_symbols[5],solve_a5).subs(all_symbols[6],solve_a6).subs(all_symbols[8],solve_a8).subs(all_symbols[9],solve_a9)
a_vec_solved
```

```python
full_vector_solution_1 = fst.OperSequence()
for i,seq in enumerate(all_seqs):
    full_vector_solution_1 += a_vec_solved[i]*seq
full_vector_solution_1.remove_zero_weight()
display(Markdown('$\\gamma$=' + full_vector_solution_1._repr_markdown_()))
```

```python
## Sanity check that the solution indeed square to 1 (requires some expanding and simping on the trig to see that some terms cancel to zero)
square = full_vector_solution_1**2
for idx,w in enumerate(square.weights):
    square.weights[idx] = sp.trigsimp(sp.expand_trig(w))
square.remove_zero_weight()
square
```

### Anticomuttation test

```python
def generate_majorana(a_vec, basis):
    assert len(a_vec) == len(basis)
    operator = fst.OperSequence()
    for i,seq in enumerate(all_seqs):
        operator += a_vec[i]*seq
    operator.remove_zero_weight()
    return operator
```

```python
## A solution seems to be maj2 with theta1=0 and theta2 = theta2+np.pi/2
## This solution does not contain the canonical one, probably because we have restricted the system to be real
maj1 = generate_majorana(a_vec_solved, all_seqs)
maj2 = generate_majorana(sp.trigsimp(a_vec_solved.subs(theta1, 0).subs(theta2,theta2+np.pi/2)), all_seqs)
```

```python
def kill_eps(expr, eps=1e-14):
    return expr.replace(
        lambda x: x.is_Number and abs(complex(x)) < eps,
        lambda x: sp.Integer(0)
    )
```

```python
def simpl_opers(oper_sequence):
    for idx,w in enumerate(oper_sequence.weights):
        oper_sequence.weights[idx] = kill_eps(sp.trigsimp(sp.expand_trig(w)))
    oper_sequence.remove_zero_weight()
```

```python
commutator = maj1*maj2 - maj2*maj1
simpl_opers(commutator)
commutator
```

```python
squared = maj1**2
simpl_opers(squared)
squared
```

```python
squared = maj2**2
simpl_opers(squared)
squared
```

## Constrained & Real

```python
all_seqs = [1, c_up, c_down, a_up, a_down, 
            c_up*a_up, c_down*a_down, c_up*a_down, c_down*a_up, 
            c_down*c_up,a_down*a_up,
            c_up*a_down*a_up, c_down*a_down*a_up, c_down*c_up*a_down, c_down*c_up*a_up,  
            c_up*c_down*a_up*a_down]

## Project the full basis onto a constrained system
P = -c_down*a_down*c_up*a_up+1
constr_seqs = []
for seq in all_seqs:
    projected = P*seq*P
    if projected == zero:
        continue
    constr_seqs.append(projected)

N_basis_vecs = len(constr_seqs)
identity_vec = sp.Matrix(N_basis_vecs, 1, lambda i, _: 0)
identity_vec[0]=1

## Check which operators are self_adjoint
self_adjoint = []
symbols = []
for seq in constr_seqs:
    conj = ~seq
    conj.normal_order()
    self_adjoint.append(conj==seq)
## Generate a test vector
full_vector = fst.OperSequence()
for i,seq in enumerate(constr_seqs):
    if self_adjoint[i]:
        a = sp.Symbol(f'a_{i}', real=True, positive=False,commutative=True)
    else:
        a = sp.Symbol(f'a_{i}', real=True, positive=False, commutative=True)
    symbols.append(a)
    
    full_vector += a*seq
```

```python
M = sp.Matrix(N_basis_vecs, N_basis_vecs, lambda i, j: 0)
for idx,symb in enumerate(symbols):
    M[idx,0]=symb
```

```python
def test_solution(paraferm):

    ### Test gamma^3 = 1
    cube = paraferm**3
    for idx,w in enumerate(cube.weights):
        cube.weights[idx] = sp.simplify(w)
    cube.remove_zero_weight()
    extra_deep_scrub(cube)
    is_unity = (cube==P)
    display(Markdown('$\\gamma^3$ = ' + cube._repr_markdown_() + f' ({is_unity})'))

    ### Test gamma^2 = gamma^dagger
    conj = ~paraferm
    conj.normal_order()
    conj
    for idx,w in enumerate(conj.weights):
        conj.weights[idx] = sp.simplify(w)
    conj.remove_zero_weight()

    sq = paraferm**2
    for idx,w in enumerate(sq.weights):
        sq.weights[idx] = sp.simplify(w)
    sq.remove_zero_weight()
    is_hermit = (sq == conj)
    display(Markdown('$\\gamma^2$ = $\\gamma^{\\dagger}$' + f' ({is_hermit})'))
```

```python
def extra_deep_scrub(op):
    for idx in range(len(op.weights)-1, -1,-1):
        if op.weights[idx].round(14).is_zero:
            op.weights.pop(idx)
            op.oper_list.pop(idx)
```

```python
for i in range(5):
    paraferm = get_solution_set1(i)[1]
    test_solution(paraferm)
```

## Matrix Approach

```python
def extra_deep_scrub(op):
    for idx in range(len(op.weights)-1, -1,-1):
        if op.weights[idx].round(14).is_zero:
            op.weights.pop(idx)
            op.oper_list.pop(idx)

# Define a threshold for "small" numbers
threshold = 1e-12

def clean_complex(z, tol=threshold):
    z = sp.expand(z)
    re = 0.0 if abs(sp.re(z)) < tol else sp.re(z)
    im = 0.0 if abs(sp.im(z)) < tol else sp.im(z)
    return complex(re, im)
    
def test_solution(paraferm):
    ### Test gamma^3 = 1
    cube = paraferm**3
    for idx,w in enumerate(cube.weights):
        cube.weights[idx] = clean_complex(sp.simplify(w))
    cube.remove_zero_weight()
    #extra_deep_scrub(cube)
    is_unity = (cube==P)
    display(Markdown('$\\chi^3$ = ' + cube._repr_markdown_() + f' ({is_unity})'))

    ### Test gamma^2 = gamma^dagger
    conj = ~paraferm
    conj.normal_order()
    conj
    for idx,w in enumerate(conj.weights):
        conj.weights[idx] = sp.simplify(sp.expand(w))
    conj.remove_zero_weight()

    sq = paraferm**2
    for idx,w in enumerate(sq.weights):
        sq.weights[idx] = sp.simplify(sp.expand(w))
    sq.remove_zero_weight()
    is_hermit = (sq == conj)
    display(Markdown('$\\chi^2$ = $\\chi^{\\dagger}$' + f' ({is_hermit})'))

def test_solutions(paraferm_vec1, paraferm_vec2):
    if not isinstance(paraferm_vec1,fst.OperSequence):
        u = make_oper(paraferm_vec1)
        v = make_oper(paraferm_vec2)
    else:
        u = paraferm_vec1
        v = paraferm_vec2
    print('Operator 1:')
    test_solution(u)
    
    print('Operator 2:')
    test_solution(v)

    comm = (u)*(v) - omega*(v)*u
    for idx,w in enumerate(comm.weights):
        comm.weights[idx] = sp.simplify(w)
    comm.remove_zero_weight()
    print('Commutator:')
    if comm == zero:
        display(Markdown('$\\chi_1\\chi_2 - \\omega\\chi_2\\chi_1=0$'))
    else:
        print('Commutation FAILED')


def test_solutions_matr(paraferm_matr1, paraferm_matr2):
    u = matrix_to_operator(paraferm_matr1)
    v = matrix_to_operator(paraferm_matr2)
    print('Operator 1:')
    test_solution(u)
    
    print('Operator 2:')
    test_solution(v)

    comm = (u)*(v) - omega*(v)*u
    for idx,w in enumerate(comm.weights):
        comm.weights[idx] = sp.simplify(sp.expand(w))
    comm.remove_zero_weight()
    print('Commutator:')
    if comm == zero:
        display(Markdown('$\\chi_1\\chi_2 - \\omega\\chi_2\\chi_1=0$'))
    else:
        print_md(comm)
        print('Commutation FAILED')
    return u,v

def test_solution_W(W):
    U_prime = sp.simplify(W.H*U*W)
    V_prime = sp.simplify(W.H*V*W)
    u,v = test_solutions_matr(U_prime,V_prime)
    return u,v

def make_oper(vector):
    oper = fst.OperSequence()
    for idx,v in enumerate(vector):
        oper+= v*constr_seqs[idx]
    oper.remove_zero_weight()
    return oper
```

```python
all_seqs = [1, c_up, c_down, a_up, a_down, 
            c_up*a_up, c_down*a_down, c_up*a_down, c_down*a_up, 
            c_down*c_up,a_down*a_up,
            c_up*a_down*a_up, c_down*a_down*a_up, c_down*c_up*a_down, c_down*c_up*a_up,  
            c_up*c_down*a_up*a_down]

## Project the full basis onto a constrained system
P = -c_down*a_down*c_up*a_up+1
constr_seqs = []
for seq in all_seqs:
    projected = P*seq*P
    if projected == zero:
        continue
    constr_seqs.append(projected)

N_basis_vecs = len(constr_seqs)
identity_vec = sp.Matrix(N_basis_vecs, 1, lambda i, _: 0)
identity_vec[0]=1

## Check which operators are self_adjoint
self_adjoint = []
symbols = []
for seq in constr_seqs:
    conj = ~seq
    conj.normal_order()
    self_adjoint.append(conj==seq)
## Generate a test vector
full_vector = fst.OperSequence()
for i,seq in enumerate(constr_seqs):
    if self_adjoint[i]:
        a = sp.Symbol(f'a_{i}', real=True, positive=False,commutative=True)
    else:
        a = sp.Symbol(f'a_{i}', real=True, positive=False, commutative=True)
    symbols.append(a)
    
    full_vector += a*seq

omega = sp.nsimplify(sp.exp(2*sp.pi*sp.I/3))
```

### Generalize Solution

```python
## Initial U and V matrices
U = sp.Matrix(3, 3, lambda i, j: 0)
U[0,1] = 1
U[1,2] = 1
U[2,0] = 1

V = sp.Matrix(3, 3, lambda i, j: 0)
V[0,0] = 1
V[1,1] = omega
V[2,2] = omega**2
```

```python
n_up = c_up*a_up
n_down = c_down*a_down
base_vec = [
    [P*(-n_up-n_down+1)*P, P*a_down*P, P*a_up*P],
     [P*c_down*P ,P*n_down*P , P*c_down*a_up*P],
     [P*c_up*P, P*c_up*a_down*P , P*n_up*P],
]
def matrix_to_operator(matrix):
    oper = fst.OperSequence()
    for i in range(3):
        for j in range(3):
            oper+= matrix[i,j]*base_vec[i][j]
    oper.remove_zero_weight()
    return oper

def show_solution(W):
    U_prime = sp.simplify(W.H*U*W)
    V_prime = sp.simplify(W.H*V*W)
    u = matrix_to_operator(U_prime)
    v = matrix_to_operator(V_prime)

    for idx,w in enumerate(u.weights):
        u.weights[idx] = sp.simplify(sp.expand(w))
    u.remove_zero_weight()

    for idx,w in enumerate(v.weights):
        v.weights[idx] = sp.simplify(sp.expand(w))
    v.remove_zero_weight()

    prod = u*v
    for idx,w in enumerate(prod.weights):
        prod.weights[idx] = sp.simplify(sp.expand(w))
    prod.remove_zero_weight()
    display(Markdown('$\\chi_1 = $' + u._repr_markdown_()))
    display(Markdown('$\\chi_2 = $' + v._repr_markdown_()))
    #display(Markdown('$\\chi_1\\chi_2 = $' + prod._repr_markdown_()))
    #print((prod)%2)
    return u,v
```

```python
def odd_even_split(oper):
    oper_list_even = []
    oper_list_odd = []
    weight_list_even = []
    weight_list_odd = []
    for idx,w in enumerate(oper.oper_list):
        if len(w) % 2:
            oper_list_odd.append(w)
            weight_list_odd.append(oper.weights[idx])
        else:
            oper_list_even.append(w)
            weight_list_even.append(oper.weights[idx])

    even_op = fst.OperSequence(oper_list_even, weights=weight_list_even,bypass_parse=True)
    odd_op = fst.OperSequence(oper_list_odd, weights=weight_list_odd,bypass_parse=True)
    return even_op, odd_op
```
